```python
from zoneinfo import ZoneInfo
from hdate import HDate, Zmanim
from datetime import date
from typing import List

# Initial cycle begin dates
cycle_begin_dates: List[HDate] = list(map(HDate, [date(2014, 12, 20), date(2018, 7, 14), date(2022, 2, 5)]))

dates: List[List[int]] = []

def advance_date(hdate_obj: HDate, steps: int) -> HDate:
    for _ in range(steps):
        hdate_obj = hdate_obj.next_day
    return hdate_obj


for i in range(929):
    steps: int = 3 if i % 5 == 0 and i != 0 else 1
    curr_dates: List[HDate] = [advance_date(hdate_item, steps) for hdate_item in (cycle_begin_dates if i == 0 else curr_dates)]
    perek_dates: List[int] = [int(f"{hdate_item.hdate.year}{str(hdate_item.hdate.month.value).zfill(2)}{str(hdate_item.hdate.day).zfill(2)}") for hdate_item in curr_dates]
    dates.append(perek_dates)

print(dates)
```
